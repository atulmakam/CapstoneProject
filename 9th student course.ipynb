# **9th student course**

This script is to get the courses that have at least 100 records.

```python
# Irene

import pandas as pd
import numpy as np
import re
```

```python
# Irene

allcourses = pd.read_feather('./Allcourses/allcourses_new.feather')
student_9th = pd.read_csv('9th_student_info.csv', index_col=0)
student_9th.index = student_9th.index.astype(np.int64)
student_9th.index.name = 'study_id'
course_info = pd.read_feather('./Allcourses/course_info.feather')
```

```python
# Irene

display(allcourses.head())
display(student_9th.head())
display(course_info.head())
```

```python
# Irene

allcourses = allcourses[['study_id', 'school_year', 'coded_course', 'ib_indicator', 'ap_indicator', 
                         'de_indicator', 'cte_indicator', 'h_indicator']].merge(course_info[['coded_course', 'course_title']], on='coded_course')
allcourses = allcourses.merge(student_9th['grade9year'], left_on='study_id', right_index=True)
allcourses = allcourses[allcourses['school_year'] == allcourses['grade9year']].drop(columns=['school_year', 'grade9year'])
allcourses
```

```python
# Irene

def replace_txt(x):
#     x = re.sub(r'\bCTE|AP|IB|DE\b', '', x)
    x = str.lower(x)
    x = re.sub(r'\bi\b', '1', x)
    x = re.sub(r'\bii\b', '2', x)
    x = re.sub(r'\biii\b', '3', x)
    x = re.sub(r'\biv\b', '4', x)
    x = re.sub(r'\bv\b', '5', x)
    x = re.sub(r'\bvi\b', '6', x)
    x = re.sub(r'\bvii\b', '7', x)
    x = re.sub(r'\bviii\b', '8', x)
    x = re.sub(r'-|\s{2,}', ' ', x)
    x = x.strip()
    return x
```

```python
# Irene

allcourses['course_title'] = allcourses['course_title'].transform(replace_txt)
```

```python
# Irene

course_count = allcourses['course_title'].value_counts()
drop_course = course_count[course_count > 100].index
allcourses = allcourses[allcourses['course_title'].isin(drop_course)].reset_index(drop=True)
allcourses.drop(columns='study_id').drop_duplicates().to_csv('./Allcourses/9th_student_course.csv', index=False)
allcourses
```
