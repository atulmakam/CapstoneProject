# **Course ID and Title**

This file is for finding the different course titles that have the same course id.

```python
# Irene

import pandas as pd
import numpy as np
```

```python
# Irene

df = pd.read_feather('course_info.feather')
```

```python
# Irene

course_id_title = df[['district','course_id', 'course_title']].copy()
course_id_title = course_id_title.groupby(['district','course_id'])['course_title'].apply(set).reset_index()
course_id_title = pd.concat([course_id_title, pd.DataFrame(course_id_title['course_title'].values.tolist())], axis=1).drop(columns='course_title')
course_id_title
```

```python
# Irene

course_id_title.to_csv('course_id_title.csv')
```
