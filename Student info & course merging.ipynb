# **Student info & course merging**

This script is to combine the student info and the course count info together.

```python
# Irene

import pandas as pd
import numpy as np
```

```python
# Irene

student_grade = pd.read_feather('./Allcourses/student_grade.feather')
student_grade.set_index('study_id', inplace=True)
student_grade.drop(columns='district', inplace=True)
student_grade
```

```python
# Irene

df = pd.read_csv('./Districts/d_9th.csv', index_col=0, low_memory=False)
df
```

```python
# Irene

student_info = pd.merge(df, student_grade, left_index=True, right_index=True)
student_info
```

```python
# Irene

# drop less2year_public because all the values are 0
student_info.drop(columns='less2year_public', inplace=True)
```

```python
student_info.columns
```

```python
# Irene

student_info.to_csv('9th_student_info.csv')
```
