# **9th student info and course**

This script is to create the files that contain the student info and their course info.

### **Summary**
1. **Read Data** [(→)](#1.-Read-Data)\
Gather the data and clean.
2. **Student Info and Courses** [(→)](#2.-Student-Info-and-Courses)\
Construct the file containing the student info and the number of each course they took.
3. **Student Info and Broader Subjects** [(→)](#3.-Student-Info-and-Broader-Subjects)\
Construct the file containing the student info and the number of courses they took that belong to broader subjects.
4. **Student Info and Narrower Subjects** [(→)](#4.-Student-Info-and-Narrower-Subjects)\
Construct the file containing the student info and the number of courses they took that belong to narrower subjects.

```python
# Irene

import pandas as pd
import numpy as np
import re
```

#### **1. Read Data** 
[(→)](#Summary)

```python
# Irene

df = pd.read_excel('./Allcourses/9th_student_course.xlsx', sheet_name='course_title')
df
```

```python
# Irene

# remove the numbers in the course title
def process_title(x):
    temp = re.sub(r'\d', '', x[-8:])
    x = x[:-8] + temp
    x = re.sub(r'  ', ' ', x)
    x = x.strip()
    return x
```

```python
# Irene

df['adj_course_title3'] = df['adj_course_title2'].transform(process_title)
df
```

```python
# Irene

# a dict that map the original course title to the adjusted course title 
replace_dict = {}
for idx in range(df.shape[0]):
    replace_dict[df.loc[idx, 'course_title']] = df.loc[idx, 'adj_course_title3']
```

```python
# Irene

allcourses = pd.read_feather('./Allcourses/allcourses_new.feather')
student_9th = pd.read_csv('9th_student_info.csv', index_col=0)
student_9th.index = student_9th.index.astype(np.int64)
student_9th.index.name = 'study_id'

course_info = pd.read_feather('./Allcourses/course_info.feather')
```

```python
# Irene

student_9th['ethnicity'].replace({'Native Hawaiian or Other Pacific Islander':'Native Hawaiian/Pacific Islander'}, inplace=True)
```

```python
# Irene

allcourses = allcourses[['study_id', 'school_year', 'coded_course']].merge(course_info[['coded_course', 'course_title']], on='coded_course')
allcourses = allcourses.merge(student_9th['grade9year'], left_on='study_id', right_index=True)
allcourses = allcourses[allcourses['school_year'] == allcourses['grade9year']].drop(columns=['school_year', 'grade9year'])
allcourses
```

```python
# Irene

# replace the Roman numerals to Arabic numerals
def replace_txt(x):
#     x = re.sub(r'\bCTE|AP|IB|DE\b', '', x)
    x = str.lower(x)
    x = re.sub(r'\bi\b', '1', x)
    x = re.sub(r'\bii\b', '2', x)
    x = re.sub(r'\biii\b', '3', x)
    x = re.sub(r'\biv\b', '4', x)
    x = re.sub(r'\bv\b', '5', x)
    x = re.sub(r'\bvi\b', '6', x)
    x = re.sub(r'\bvii\b', '7', x)
    x = re.sub(r'\bviii\b', '8', x)
    x = re.sub(r'-|\s{2,}', ' ', x)
    x = x.strip()
    return x
```

```python
# Irene

allcourses['course_title'] = allcourses['course_title'].transform(replace_txt)
allcourses
```

```python
# Irene

# keep only the courses that have at least 100 records
course_count = allcourses['course_title'].value_counts()
drop_course = course_count[course_count > 100].index
allcourses = allcourses[allcourses['course_title'].isin(drop_course)].reset_index(drop=True)
allcourses = allcourses.drop(columns='coded_course')
allcourses
```

```python
# Irene

allcourses_ori = allcourses.copy()
```

---

#### **2. Student Info and Courses** 
[(→)](#Summary)

```python
# Irene

allcourses['course_title'].replace(replace_dict, inplace=True)
allcourses
```

```python
# Irene

len(allcourses['course_title'].unique())
```

```python
# Irene

student_course = allcourses.groupby('study_id')['course_title'].apply(set)#.set_index('study_id')
student_course =  pd.DataFrame(student_course.values.tolist(), index=student_course.index)
student_course
```

```python
# Irene

course_col = pd.Series(list(set(student_course.values.flatten()))).sort_values().dropna()
student_course_dummy = pd.DataFrame(0, columns=course_col, index=student_course.index)
student_course_dummy_ori = student_course_dummy.copy()
student_course_dummy
```

```python
# Irene

def fill_dummy(x):
    student_course_dummy.loc[x.name, x[x.notna()].values] = 1
student_course.apply(fill_dummy, axis=1)
student_course_dummy
```

```python
# Irene

student_9th.drop(columns='birth_date', inplace=True)
student_9th_course = pd.merge(student_9th, student_course_dummy, left_index=True, right_index=True)
student_9th_course
```

```python
# Irene

student_9th_course[student_9th_course.iloc[:,4:12].columns].astype(bool)
bool_list = list(student_9th_course.iloc[:,4:12].columns)
bool_list.extend(list(student_9th_course.iloc[:,47:].columns))
bool_list.remove('free_reduced_lunch_indicator')
student_9th_course = student_9th_course.astype({k:bool for k in bool_list})
notna_idx = student_9th_course[student_9th_course['free_reduced_lunch_indicator'].notna()].index
student_9th_course.loc[notna_idx, 'free_reduced_lunch_indicator'] = student_9th_course.loc[notna_idx, 'free_reduced_lunch_indicator'].transform(lambda x: True if x == 1 else False)
student_9th_course['grade9year'] = student_9th_course['grade9year'].astype(np.int16)
student_9th_course['GPA'] = student_9th_course['GPA'].astype(np.float16)
student_9th_course.sort_values(by=['district', 'study_id'], inplace=True)
student_9th_course
```

```python
# Irene

student_9th_course.drop(columns=['del'], inplace=True)
student_9th_course.to_csv('9th_student_info_course2.csv', index_label='study_id')
```

#### **3. Student Info and Broader Subjects** 
[(→)](#Summary)

```python
# Irene

allcourses = allcourses_ori.copy()
allcourses
```

```python
# Irene

allcourses_subject = allcourses.merge(df[['course_title', 'subject']], on='course_title')
allcourses_subject.drop(columns='course_title', inplace=True)
allcourses_subject
```

```python
# Irene

allcourses_subject['subject2'] = allcourses_subject['subject']
student_subject = allcourses_subject.pivot_table(values='subject2', index='study_id', columns='subject', aggfunc=pd.Series.count)
student_subject = student_subject.fillna(0)
student_subject
```

```python
# Irene

student_9th_subject = pd.merge(student_9th, student_subject, left_index=True, right_index=True)
student_9th_subject
```

```python
# Irene

student_9th_subject[student_9th_subject.iloc[:,4:12].columns].astype(bool)
bool_list = list(student_9th_subject.iloc[:,4:12].columns)
int_list = list(student_9th_subject.iloc[:,47:].columns)
bool_list.remove('free_reduced_lunch_indicator')
student_9th_subject = student_9th_subject.astype({k:bool for k in bool_list})
student_9th_subject = student_9th_subject.astype({k:np.int8 for k in int_list})
notna_idx = student_9th_subject[student_9th_subject['free_reduced_lunch_indicator'].notna()].index
student_9th_subject.loc[notna_idx, 'free_reduced_lunch_indicator'] = student_9th_subject.loc[notna_idx, 'free_reduced_lunch_indicator'].transform(lambda x: True if x == 1 else False)
student_9th_subject['grade9year'] = student_9th_subject['grade9year'].astype(np.int16)
student_9th_subject['GPA'] = student_9th_subject['GPA'].astype(np.float16)
student_9th_subject.sort_values(by=['district', 'study_id'], inplace=True)
student_9th_subject
```

```python
# Irene

student_9th_subject.drop(columns=['del'], inplace=True)
student_9th_subject.to_csv('9th_student_info_course3.csv', index_label='study_id')
```
