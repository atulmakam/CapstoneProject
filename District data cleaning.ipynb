# **District data cleaning**

This script is to clean and merge the 5 district files. We check the null values, correct the typos and impute missing values if necessary.

---

### **Summary**
1. **Explore Data and Classify Columns** [(→)](#1.-Explore-Data-and-Classify-Columns)\
Have a look at the data, drop some unnecessary data, and classify the columns (e.g. college or high school related). 
2. **Get Student's Basic Info** [(→)](#2.-Get-Student's-Basic-Info)\
Create dataframes named "student_info" for each district to store the desired info for output purpose. Find the 9th grade year of the students, and their basic info and indicators (e.g. high_school_grad, college_grad).
3. **Get Student's High School Info** [(→)](#3.-Get-Student's-High-School-Info)\
Get student's high school related info (high school code).
4. **Get Student's College Info** [(→)](#4.-Get-Student's-College-Info)\
Get student's college related info (e.g. degree, college type).
5. **Drop Students and Save to Files** [(→)](#5.-Drop-Students-and-Save-to-Files)\
Drop the students who doesn't have any college info (like some students was 9th grade in 2006 while the collected college data starts from 2011). Then save the dataframes separately by districts and also to a combined file.

---

```python
# Irene

# find the grade_level info for each district
for df_name, df in df_l:
    df_d[df_name]['grade_level'] = df_d[df_name]['high_school'].filter(regex=re.compile(r'FY\d{4}grade_level'), axis=1).copy()
    print(list(df_d[df_name]['grade_level'].columns))
```

```python
# Irene

# a function that can calculate the 9th grade year of students if they have at least one grade_level record
# -1 will be returned if there's no grade_level record
def year_level(x):
    if x.notna().sum() == 0:
        return -1
    else:
        grade = x[x.notna()].copy()
        if 9 in grade.values:
            grade9year = list(grade.index[grade.values==9])[0][2:6]
            return grade9year
        else:
            grade = sorted(tuple(zip(grade.index, grade.values)), key=lambda x: (x[0], x[1]))[0]
            grade9year = int(grade[0][2:6]) - (grade[1] - 9)
            return grade9year
```

```python
# Irene

# student_info will store all the information needed for regression analysis
# find the 9th grade year of every student
for df_name, df in df_l:
    dfd = df_d[df_name]
    dfd['student_info'] = pd.DataFrame(index=dfd['grade_level'].index)
    dfd['student_info']['grade9year'] = dfd['grade_level'].apply(year_level, axis=1).astype(np.int16)
    if df_name == 'df4':
        print(df_name)
        dfd['student_info']['grade9year'] = dfd['student_info']['grade9year'].transform(lambda x: x + 1 if x != -1 else x)
    display(dfd['student_info'].head())
    print(dfd['student_info'][dfd['student_info'].grade9year==-1].shape)
```

```python
# Irene

# a function that can impute the 9th grade year based on birthdate
# if birth month is greater than 9, the 9th grade year will be birth year + 15, else will be + 14
def birthdate_impute(x):
    return_list = []
    for i in x:
        m = pd.DatetimeIndex([i]).month
        y = pd.DatetimeIndex([i]).year
        if m >= 9:
            return_list.append(y + 15)
        else:
            return_list.append(y + 14)
    return return_list
```

```python
# Irene

# add some basic information about the students and indicators to student_info
for df_name, df in df_l:
    dfd = df_d[df_name]
    
    student_info = dfd['student_info']
    college_grad = dfd['college']
    
    student_info[dfd['info'].columns[0:-1]] = dfd['info'].iloc[:, 0:-1]
    
    # distinguish if the student graduated from high school
    # 1 = graduated from high school, 0 = didn't graduate or doesn't have data (like maybe transfer to other state)
    student_info['high_school_grad'] = dfd['info']['high_school_grad_date'].notna().transform(lambda x: 1 if x == True else 0)
    grad_index = student_info[student_info['high_school_grad']==1].index
    
    # for gradelevel is unknown, find the graduate date or birth date and calualate which year the student was on grade 9
    unknown9grade_index = student_info[student_info['grade9year'] == -1].index
    unknown9grade = dfd['info'].loc[unknown9grade_index]
    impute9grade_index = unknown9grade[unknown9grade['high_school_grad_date'].notna()].index
    student_info.loc[impute9grade_index, 'grade9year'] = pd.DatetimeIndex(dfd['info'].loc[impute9grade_index, 'high_school_grad_date']).year - 3
    
    unknown9grade_index = student_info[student_info['grade9year'] == -1].index
    unknown9grade = dfd['info'].loc[unknown9grade_index]
    impute9grade_index = unknown9grade[unknown9grade['birth_date'].notna()].index
    student_info.loc[impute9grade_index, 'grade9year'] = dfd['info'].loc[impute9grade_index, 'birth_date'].transform(birthdate_impute)
    
    # distinguish if the student is still studying in high school
    # 1 = still in high school, 0 = not in high school
    student_info['high_school_studying'] = 0 
    not_grad_index = student_info[~(student_info.index.isin(grad_index))].index
    high_school_studying_index = dfd['info'].loc[not_grad_index][student_info.loc[not_grad_index, 'grade9year'] + 3 > dfd['latest_year']].index
    student_info.loc[high_school_studying_index, 'high_school_studying'] = 1

    
    # distinguish if the student enrolled in college
    student_info['college_enroll'] = 0
    student_info['college_enroll'] = dfd['college'].filter(regex=re.compile(r'FY\d{4}college\d.*')).notna().sum(axis=1).transform(lambda x: 1 if x > 0 else 0)
    
    # distinguish if the student graduated from college
    student_info['college_graduate'] = dfd['college'].filter(regex=re.compile(r'college_grad.*')).notna().sum(axis=1).transform(lambda x: 1 if x > 0 else 0)

    # distinguish if the student is still studying in college
    # this part will be processed in later cell
    student_info['college_studying'] = 0
    
    if df_name not in ['df1', 'df2', 'df5']:
#         student_info['free_reduced_lunch_indicator'] = student_info['free_reduced_lunch_indicator'].fillna(-1).astype(np.int8)
        pass
    else:
        student_info['free_reduced_lunch_indicator'] = student_info['free_reduced_lunch_indicator'].fillna(0).astype(np.int8)
        
    display(student_info.head())
```

```python
# Irene

# find the high school code when the student was on 9th grade level
# if there's no code data at all, return nan
# if the student has high school code data but it's not the 9th grade year one, return the first record of code
def get_high_school_code(x):
    return_value = None
    codes = x[1:][x[1:].notna()]
    if len(codes) == 0:
        return np.nan
    else:
        y = x['grade9year']
        ycodes = codes.filter(regex=re.compile(str(y)))
        ycodes = ycodes[ycodes.notna()]
        if len(ycodes) >= 1:
            return ycodes.values[0]
        else:
            return codes.values[0]
```

```python
# Irene

# find high_school_code and store it to student_info
for df_name, df in df_l:
    dfd = df_d[df_name]
    student_info = dfd['student_info']
    high_school_code = dfd['high_school'].filter(regex=re.compile(r'.*high_school_code.*'))
    high_school_code = pd.concat([dfd['student_info']['grade9year'], high_school_code], axis=1)
    student_info['high_school_code'] = high_school_code.apply(get_high_school_code, axis=1)
    display(student_info.head())
```

```python
# Irene

# list the college graduation related columns
for df_name, df in df_l:
    college_grad = df_d[df_name]['college'].filter(regex=re.compile(r'^college.*'))
    print(college_grad.columns)
```

```python
# Irene

# a function that can find the highest degree that the students get
# degree_type = ['0 UNKNOWN', '1 CERTIFICATE', '2 ASSOCIATES', '3 BACHELORS',  '4 PROFESSIONAL', '5 MASTERS', '6 DOCTORATE']
def find_degree(x):
    x = sorted(x[x.notna()])
    if len(x) == 0:
        return np.nan
    else:
        return x[-1]
```

```python
# Irene

# a function that can find the first degree that the students get
def first_degree(x):
    min_idx = None
    min_date = '2099-1-1' 	
    for idx, date in x.items():
        if idx[-1] == 'e':
            continue
        if date != None and date < min_date:
            min_date = date
            min_idx = idx
    if min_idx == None:
        return None
    else:   
        return 'college_grad_degree' + min_idx[-1] + 'type'
```

```python
# Irene

# add college graduation data to student_info

# grad_cols = list(df_d['df4']['college'].filter(regex=re.compile(r'^college((?!date|detail|graduated).)*$')))

for df_name, df in df_l:
    dfd = df_d[df_name]
    college_grad = dfd['college'].filter(regex=re.compile(r'^college((?!detail|graduated).)*$')).copy()
    grad_concat = pd.DataFrame(columns=['num_degree', 'highest_degree', 'first_degree'], index=college_grad.index)
    
    # find the number of degrees the students get
    grad_concat['num_degree'] = college_grad.filter(regex=re.compile(r'name')).notna().sum(axis=1)
    
    if df_name != 'df2':
        for i in range(0, len(college_grad.columns), 3):           
            miss_degree_idx = college_grad[(college_grad.iloc[:, i+2].notna()) & (college_grad.iloc[:, i+1].isna())].index
            college_grad.loc[miss_degree_idx, college_grad.columns[i+1]] = '0 UNKNOWN'
            
        grad_concat['highest_degree'] = college_grad.filter(regex=re.compile(r'degree')).apply(find_degree, axis=1)
        
        degree_date_type = college_grad.filter(regex=re.compile(r'date|type'))
        locator = degree_date_type.apply(first_degree, axis=1)
        first_degree_l = []
        for i in degree_date_type.index:
            if locator.loc[i] == None:
                first_degree_l.append(None)
            else:
                first_degree_l.append(degree_date_type.loc[i, locator.loc[i]])        
        grad_concat['first_degree'] = first_degree_l
    else:
        name_df = college_grad.filter(regex=re.compile(r'name')).copy()
        miss_degree_idx = name_df[name_df.notna().sum(axis=1) > 0].index
        grad_concat.loc[miss_degree_idx, 'first_degree'] = '0 UNKNOWN'
        grad_concat.loc[miss_degree_idx, 'highest_degree'] = '0 UNKNOWN'
        
    dfd['student_info'] = pd.concat([dfd['student_info'], grad_concat], axis=1)
    display(dfd['student_info'].head())
```

```python
# Irene

# Less Than 2 Years
def college_type(x, string):
    counter = Counter(x)
    count = 0
    if string == '4year_public':
        return counter[('4-year', 'Public')]
    elif string == '2year_public':
        return counter[('2-year', 'Public')]
    elif string == 'less2year_public':
        return counter[('Less Than 2 Years', 'Public')]
    elif string == '4year_private':
        return counter[('4-year', 'Private')]
    elif string == '2year_private':
        return counter[('2-year', 'Private')]
    elif string == 'less2year_private':
        return counter[('Less Than 2 Years', 'Private')]
    elif string == '4year':
        return college_year(counter, '4')
    elif string == '2year':
        return college_year(counter, '2')
    elif string == 'less2year':
        return college_year(counter, 'L')  
```

```python
# Irene

def college_year(counter, key):
    count = 0
    for k, v in counter.items():
        if k == None:
            continue
        elif k[0][0] == key:
            count += v
    return count
```

```python
# Irene

# add enrolled college info to student_info
for df_name, df in df_l:
    dfd = df_d[df_name]
    student_info = dfd['student_info']
    college_info = dfd['college'].filter(regex=re.compile(r'^(?!college_grad)'))
                
    extract_d1 = {id:[] for id in college_info.index}
    extract_d2 = {id:[] for id in college_info.index}
    for i in range(0, len(college_info.columns), 6):
        
        # for finding the last enrollment status
        segment1 = college_info.iloc[:, i+1:i+4]
        segment1_notna = segment1[segment1.iloc[:,2].notna()]
        segment1_dict = segment1_notna.T.to_dict('list')
        
        for id in segment1_dict.keys():
            values = tuple(segment1_dict[id])
            if values not in extract_d1[id]:
                extract_d1[id].append(values)
        
        # for finding all the enrolled college
        segment2 = college_info.iloc[:, i+4:i+6]
        segment2_notna = segment2[segment2.notna().sum(axis=1) > 0]
        segment2_dict = segment2_notna.T.to_dict('list')
        
        for id in segment2_dict.keys():
            values = tuple(segment2_dict[id])
            if values not in extract_d2[id]:
                extract_d2[id].append(values)
                
    # for finding the last enrollment status
    college_studying = []
    for id in extract_d1.keys():
        try:
            if df_name == 'df2':
                month = 4
            else:
                month = 11
            if (pd.DatetimeIndex([extract_d1[id][-1][0]]).year[0] == dfd['latest_year']) \
                    & (pd.DatetimeIndex([extract_d1[id][-1][0]]).month[0] >= month) \
                    & (extract_d1[id][-1][1][0] != '6'):
                college_studying.append(1)
            else:
                college_studying.append(0)
        except:
            college_studying.append(0)
    student_info['college_studying'] = college_studying
    grad_idx = student_info[student_info['college_graduate']==1].index
    student_info.loc[grad_idx, 'college_studying'] = 0

    # for finding all the enrolled college
    college_df = pd.DataFrame.from_dict(extract_d2, orient='index')
    
    college_concat = pd.DataFrame(index=college_df.index, 
                                  columns=['4year', '2year', 'less2year', 
                                           '4year_public', '4year_private', 
                                           '2year_public', '2year_private', 
                                           'less2year_public', 'less2year_private']).fillna(0)
    for c in college_concat:
        college_concat[c] = college_df.apply(college_type, axis=1, string=c)

    dfd['student_info'] = pd.concat([dfd['student_info'], college_concat], axis=1)
    display(dfd['student_info'].head())
```
